```python
#Evaluation Phase 
import json
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from collections import defaultdict

# ================================
# LOAD EMBEDDED CHUNKS
# ================================

EMBEDDED_JSON = r"C:\Users\kriti\OneDrive\Desktop\Infosys\chunks\all_visa_chunks_with_embeddings.json"

with open(EMBEDDED_JSON, "r", encoding="utf-8") as f:
    chunks = json.load(f)

print(f"Loaded {len(chunks)} embedded chunks\n")

# ================================
# PREPARE DATA
# ================================

embeddings = np.array([c["embedding"] for c in chunks])
visa_types = [c["visa_type"] for c in chunks]

# ================================
# SIMILARITY MATRIX
# ================================

sim_matrix = cosine_similarity(embeddings, embeddings)

# ================================
# METRIC COMPUTATION
# ================================

def evaluate_retrieval(k_values=[1, 3, 5]):
    recall = {k: 0 for k in k_values}
    mrr_total = 0
    pos_sims = []
    neg_sims = []

    for i in range(len(chunks)):
        sims = sim_matrix[i]
        ranked_indices = np.argsort(-sims)

        ranked_indices = ranked_indices[ranked_indices != i]  # remove self

        # --- Recall@K ---
        for k in k_values:
            top_k = ranked_indices[:k]
            if any(visa_types[j] == visa_types[i] for j in top_k):
                recall[k] += 1

        # --- MRR ---
        rank = None
        for idx, j in enumerate(ranked_indices, start=1):
            if visa_types[j] == visa_types[i]:
                rank = idx
                break
        if rank:
            mrr_total += 1 / rank

        # --- Similarity stats ---
        pos_sims.append(sim_matrix[i][i])  # self similarity = 1
        neg_sims.extend(
            sim_matrix[i][j] for j in ranked_indices[:10]
            if visa_types[j] != visa_types[i]
        )

    n = len(chunks)
    recall = {k: v / n for k, v in recall.items()}
    mrr = mrr_total / n

    return recall, mrr, np.mean(pos_sims), np.mean(neg_sims)


recall, mrr, pos_sim, neg_sim = evaluate_retrieval()

# ================================
# PRINT REPORT
# ================================

print("ðŸ“Š EMBEDDING PERFORMANCE REPORT")
print("================================\n")

for k, v in recall.items():
    print(f"Recall@{k}: {v*100:.2f}%")

print(f"\nMRR: {mrr:.3f}")

print(f"\nAverage Positive Similarity (self): {pos_sim:.3f}")
print(f"Average Negative Similarity:        {neg_sim:.3f}")

print("\n================================")
print("âœ… Embedding evaluation completed")
```
